```python
import pandas as pd

df = pd.read_csv("../data/processed/cleaned_online_retail.csv")
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])

df.head()
```

```python
recency = (
    df.groupby("Customer ID")["InvoiceDate"]
      .max()
      .reset_index()
)

recency["Recency"] = (snapshot_date - recency["InvoiceDate"]).dt.days
```

```python
frequency = (
    df.groupby("Customer ID")["Invoice"]
      .nunique()
      .reset_index()
)

frequency.columns = ["Customer ID", "Frequency"]
```

```python
df["TotalPrice"] = df["Quantity"] * df["Price"]

monetary = (
    df.groupby("Customer ID")["TotalPrice"]
      .sum()
      .reset_index()
)

monetary.columns = ["Customer ID", "Monetary"]
```

```python
rfm = recency.merge(frequency, on="Customer ID")
rfm = rfm.merge(monetary, on="Customer ID")

rfm = rfm.drop(columns=["InvoiceDate"])
rfm.head()
```

```python
# Average Order Value
rfm["AvgOrderValue"] = rfm["Monetary"] / rfm["Frequency"]

# Unique products
unique_products = (
    df.groupby("Customer ID")["StockCode"]
      .nunique()
      .reset_index()
)

unique_products.columns = ["Customer ID", "UniqueProducts"]

rfm = rfm.merge(unique_products, on="Customer ID")
```

```python
last_purchase = (
    df.groupby("Customer ID")["InvoiceDate"]
      .max()
      .reset_index()
)

last_purchase["Churn"] = (
    (snapshot_date - last_purchase["InvoiceDate"]).dt.days > 90
).astype(int)

rfm = rfm.merge(
    last_purchase[["Customer ID", "Churn"]],
    on="Customer ID"
)
```

```python
rfm.to_csv(
    "../data/processed/customer_features.csv",
    index=False
)
```

```python
# Total quantity purchased per customer
qty_features = df.groupby("Customer ID")["Quantity"].agg(
    TotalQuantity="sum",
    AvgQuantity="mean",
    MaxQuantity="max",
    MinQuantity="min"
).reset_index()

rfm = rfm.merge(qty_features, on="Customer ID")
```

```python
rfm.shape
```

```python
# Basket size per invoice
basket = (
    df.groupby(["Customer ID", "Invoice"])["Quantity"]
      .sum()
      .reset_index()
)

basket_features = basket.groupby("Customer ID")["Quantity"].agg(
    AvgBasketSize="mean",
    MaxBasketSize="max",
    MinBasketSize="min"
).reset_index()

rfm = rfm.merge(basket_features, on="Customer ID")
```

```python
rfm.shape
```

```python
# Customer purchase span and activity duration
date_features = df.groupby("Customer ID")["InvoiceDate"].agg(
    FirstPurchase="min",
    LastPurchase="max"
).reset_index()

date_features["PurchaseSpanDays"] = (
    date_features["LastPurchase"] - date_features["FirstPurchase"]
).dt.days

date_features["ActiveMonths"] = (
    (date_features["LastPurchase"].dt.to_period("M") -
     date_features["FirstPurchase"].dt.to_period("M")).apply(lambda x: x.n) + 1
)

rfm = rfm.merge(
    date_features[["Customer ID", "PurchaseSpanDays", "ActiveMonths"]],
    on="Customer ID"
)
```

```python
rfm.shape
```

```python
monetary_stats = df.groupby("Customer ID")["TotalPrice"].agg(
    AvgOrderValueTxn="mean",
    MaxOrderValueTxn="max",
    StdOrderValueTxn="std"
).reset_index()

rfm = rfm.merge(monetary_stats, on="Customer ID")
```

```python
rfm.shape
```

```python
product_features = df.groupby("Customer ID").agg(
    UniqueProducts=("StockCode", "nunique"),
    UniqueInvoices=("Invoice", "nunique")
).reset_index()

product_features["ProductsPerInvoice"] = (
    product_features["UniqueProducts"] / product_features["UniqueInvoices"]
)

rfm = rfm.merge(product_features, on="Customer ID")
```

```python
rfm.shape
```

```python
# Create UniqueProducts feature
unique_products = (
    df.groupby("Customer ID")["StockCode"]
      .nunique()
      .reset_index()
)

unique_products.columns = ["Customer ID", "UniqueProducts"]

rfm = rfm.merge(unique_products, on="Customer ID", how="left")
```

```python
rfm.columns
```

```python
# Purchases per month
rfm["PurchasesPerMonth"] = rfm["Frequency"] / rfm["ActiveMonths"]

# Monetary per product
rfm["MonetaryPerProduct"] = rfm["Monetary"] / rfm["UniqueProducts"]

# Monetary per invoice
rfm["MonetaryPerInvoice"] = rfm["Monetary"] / rfm["Frequency"]

# Recency bucket
rfm["RecencyBucket"] = pd.qcut(
    rfm["Recency"], q=4, labels=[1, 2, 3, 4]
)

# Frequency bucket
rfm["FrequencyBucket"] = pd.qcut(
    rfm["Frequency"].rank(method="first"),
    q=4, labels=[1, 2, 3, 4]
)

# Monetary bucket
rfm["MonetaryBucket"] = pd.qcut(
    rfm["Monetary"].rank(method="first"),
    q=4, labels=[1, 2, 3, 4]
)
```

```python
import pandas as pd
import json

# Load RAW dataset
df = pd.read_excel(
    "../data/raw/online_retail_II.xlsx",
    sheet_name="Year 2009-2010"
)

# Basic stats
total_rows = df.shape[0]
total_columns = df.shape[1]

# Missing values
missing_values = df.isna().sum().to_dict()

# Duplicate rows
duplicate_rows = df.duplicated().sum()

# Date range
date_range = {
    "start": str(df["InvoiceDate"].min().date()),
    "end": str(df["InvoiceDate"].max().date())
}

# Negative quantities
negative_quantities = (df["Quantity"] < 0).sum()

# Cancelled invoices (Invoice starting with 'C')
cancelled_invoices = df["Invoice"].astype(str).str.startswith("C").sum()

# Missing customer IDs
missing_customer_ids = df["Customer ID"].isna().sum()
missing_customer_ids_percentage = round(
    (missing_customer_ids / total_rows) * 100, 2
)

# Final summary dictionary
summary = {
    "total_rows": int(total_rows),
    "total_columns": int(total_columns),
    "missing_values": missing_values,
    "duplicate_rows": int(duplicate_rows),
    "date_range": date_range,
    "negative_quantities": int(negative_quantities),
    "cancelled_invoices": int(cancelled_invoices),
    "missing_customer_ids": int(missing_customer_ids),
    "missing_customer_ids_percentage": missing_customer_ids_percentage
}

# Save JSON file
with open("../data/raw/data_quality_summary.json", "w") as f:
    json.dump(summary, f, indent=2)

summary
```

```python
import pandas as pd
import json

# Load final feature dataset
df = pd.read_csv("../data/processed/customer_features.csv")

# Separate features and target
feature_cols = [c for c in df.columns if c not in ["Customer ID", "Churn"]]

# Churn stats
churn_rate = round(df["Churn"].mean(), 2)
training_customers = df.shape[0]
churned_customers = int(df["Churn"].sum())
active_customers = int(training_customers - churned_customers)

# Feature categories (based on your design)
feature_categories = {
    "rfm": 4,          # Recency, Frequency, Monetary, AvgOrderValue
    "behavioral": 8,   # basket, quantity, invoice behavior
    "temporal": 5,     # recency span, active months, purchase span
    "product": 6,      # unique products, invoices, diversity
    "derived": len(feature_cols) - (4 + 8 + 5 + 6)
}

# Build feature-level info (sample for first feature)
feature_details = []

for col in feature_cols:
    feature_details.append({
        "name": col,
        "type": "numeric",
        "description": f"{col} feature",
        "min": float(df[col].min()),
        "max": float(df[col].max()),
        "mean": float(round(df[col].mean(), 2))
    })

# Final dictionary
feature_info = {
    "total_features": len(feature_cols),
    "feature_categories": feature_categories,
    "churn_rate": churn_rate,
    "training_cutoff": "2010-09-09",
    "observation_end": "2010-12-09",
    "training_customers": training_customers,
    "churned_customers": churned_customers,
    "active_customers": active_customers,
    "features": feature_details
}

# Save JSON
with open("../data/processed/feature_info.json", "w") as f:
    json.dump(feature_info, f, indent=2)

feature_info
```
